```python
#Initialize an empty dictionary to store student records
#Format: {student_name: score}
students = {}

# Dictionary to store the grading scale for reference
grading_scale = {
    (70, 100): 'A',
    (60, 69): 'B',
    (50, 59): 'C',
    (45, 49): 'D',
    (0, 44): 'F'
}
```

```python
students = {}
```

```python
#Grading

def calculate_grade(score):
    "Return letter grade for a numeric score"
    if 70 <= score <= 100:
        return "A"
    elif 60 <= score <= 69:
        return "B"
    elif 50 <= score <= 59:
        return "C"
    elif 45 <= score <= 49:
        return "D"
    elif 0 <= score <= 45:
        return "F"
    else:
        return "Invalid Score"
```

```python
def validate_score(score):   #this is to ensure that even if the students grade is not an int, we can convert it, also check if it is between 0 - 100, 
    try:                     #returns True if valid and False if otherwise 
        score = float(score)
        return 0 <= score <= 100
    except ValueError:
        return False

def get_valid_score():   #this keeps asking for a score until a valid one is entered, then the score is returned
    while True:
        try:
            score = float(input("Enter score (0-100): "))
            if 0 <= score <= 100:
                return score
            else:
                print("Error: Score must be between 0 and 100. Please try again.")
        except ValueError:
            print("Error: Please enter a valid number.")
```

```python
def add_student():    #Adding a new student to the system 
    print("\n" + "="*40)
    print("ADD NEW STUDENT")
    print("="*40)
    
    name = input("Enter student name: ").strip()
    
    # Check if student already exists
    if name in students:
        print(f"Error: Student '{name}' already exists!")
        update = input("Would you like to update their score instead? (yes/no): ").lower()
        if update == 'yes':
            update_student_score(name)
        return
    
    score = get_valid_score()
    students[name] = score
    grade = calculate_grade(score)
    
    print(f" Student '{name}' added successfully!")
    print(f"   Score: {score} | Grade: {grade}")
```

```python
#to display all students with their scores and grades

def view_all_students():                          
    print("\n" + "="*35)                          #to just make the DB System fine
    print("ALL STUDENTS")
    print("="*35)
    
    if not students:
        print("Oopse! No students found in the system.")
        return
    
    print(f"{'Name':<20} {'Score':<10} {'Grade':<10}")
    print("-"*35)                             ##to just make the DB System fine
    
    for name, score in students.items():
        grade = calculate_grade(score)
        print(f"{name:<20} {score:<10} {grade:<10}")
    
    print("="*35)
    print(f"Total Students: {len(students)}")        #to print the total number of students
```

```python
#to update the scor of an existing student

def update_student_score(search_name = None):       
    print("\n" + "="*40)
    print("UPDATE STUDENT SCORE")
    print("="*40)
    
    if search_name is None:
        name = input("Enter student name to update: ").strip()
    else:
        name = search_name
    
    if name not in students:
        print(f" Sorry! : Student '{name}' not found!")
        return
    
    current_score = students[name]
    current_grade = calculate_grade(current_score)
    
    print(f"\nCurrent details for {name}:")
    print(f"Score: {current_score} | Grade: {current_grade}")
    
    print("\nEnter new score:")
    new_score = get_valid_score()
    
    students[name] = new_score
    new_grade = calculate_grade(new_score)
    
    print(f"\n Score updated successfully!")
    print(f"   {name}: {current_score} ({current_grade}) → {new_score} ({new_grade})")
```

```python
#to Search for a student by name and display their details. 

def search_student():      
    print("\n" + "="*40)
    print("SEARCH STUDENT")
    print("="*40)
    
    name = input("Enter student name to search: ").strip()
    
    if name in students:
        score = students[name]
        grade = calculate_grade(score)
        print("\n Student Found!")
        print("-"*30)
        print(f"Name:  {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")
        print("-"*30)
    else:
        print(f"Sorry! Student '{name}' not found in the system.")
```

```python
#to display the students records

def display_statistics():
    print("\n" + "="*50)
    print("CLASS PERFORMANCE STATISTICS")
    print("="*50)
    
    if not students:
        print("Oopse! No students in the system to calculate statistics.")
        return
    
    scores = list(students.values())
    
    highest = max(scores)
    lowest = min(scores)
    average = sum(scores) / len(scores)
    total = len(students)
    
    # Find students with highest and lowest scores
    highest_students = [name for name, score in students.items() if score == highest]
    lowest_students = [name for name, score in students.items() if score == lowest]
    
    # Grade distribution
    grade_distribution = {'A': 0, 'B': 0, 'C': 0, 'D': 0, 'F': 0}
    for score in scores:
        grade = calculate_grade(score)
        grade_distribution[grade] += 1
    
    print(f"\n Key Statistics:")
    print("-"*40)
    print(f" Highest Score: {highest}")
    print(f"   Students: {', '.join(highest_students)}")
    print(f"\n Lowest Score: {lowest}")
    print(f"   Students: {', '.join(lowest_students)}")
    print(f"\n Average Score: {average:.2f}")
    print(f" Total Students: {total}")
    
    print(f"\n Grade Distribution:")
    print("-"*40)
    for grade in ['A', 'B', 'C', 'D', 'F']:
        count = grade_distribution[grade]
        percentage = (count / total) * 100
        print(f"{grade}: {count:2} students ({percentage:5.1f}%)")
    
    print("="*50)
```

```python
display_statistics()
```

# INVENTORY MANAGEMENT SYSTEM

```python
# Initialize an empty dictionary to store products and their quantities
inventory = {}

def display_menu():                 #we want to display the main menu options so we will at a glance, see what we are working with
    print("\n" + "="*30)            #as usual, this is to make the system look fancy
    print("INVENTORY MANAGEMENT SYSTEM")
    print("="*30)
    print("1. Add Product")
    print("2. View Inventory")
    print("3. Update Quantity")
    print("4. Search Product")
    print("5. Remove Product")
    print("6. Low Stock Report")
    print("7. Inventory Statistics")
    print("8. Exit")
    print("="*30)
```

```python
#Adding a new product to the inventory 

def add_product():
    print("\n Add New Product ")
    
    # Get product name with validation (not empty)
    while True:
        product_name = input("Enter product name: ").strip().title()
        
        if not product_name:
            print("Sorry: Product name cannot be empty! Please try again.")
            continue
        
        # Check if product already exists
        if product_name in inventory:
            print(f" '{product_name}' already exists with quantity {inventory[product_name]}")
            choice = input("Do you want to update the quantity instead? (yes/no): ").lower()
            if choice == 'yes' or choice == 'y':
                # Redirect to update function
                update_quantity()
            else:
                print("Product not added.")
            return  # Exit add_product function
        
        break  # Valid product name that doesn't exist
    
    # Get quantity with comprehensive validation
    while True:
        try:
            quantity_input = input("Enter quantity: ").strip()
            
            # Check if input is empty
            if not quantity_input:
                print("Error: Quantity cannot be empty! Please enter a number.")
                continue
            
            # Try to convert to integer
            quantity = int(quantity_input)
            
            # Check for negative numbers
            if quantity < 0:
                print("Error: Quantity cannot be negative! Please enter zero or a positive number.")
                continue
            
            # Valid quantity received
            inventory[product_name] = quantity
            print(f" '{product_name}' added successfully with quantity {quantity}")
            break
            
        except ValueError:
            # This catches non-numeric inputs (alphabets, special characters)
            print("Sorry: Invalid input! Please enter a valid number (e.g., 0, 5, 10).")
            print(f"   You entered: '{quantity_input}' - This is not a number.")
```

```python
def view_inventory():             #To display all products and their quantities
    print("\n Current Inventory ")
    
    if not inventory:
        print("Inventory is empty!")
        return
    
    print("\n{:<30} {:>10}".format("Product Name", "Quantity"))
    print("-" * 42)
    
    # Sort alphabetically for better readability
    for product, quantity in sorted(inventory.items()):
        print("{:<30} {:>10}".format(product, quantity))
    print("-" * 42)
    print(f"Total unique products: {len(inventory)}")
```

```python
view_inventory()
```

```python
#Update the quantity of an existing product"""

def update_quantity():
    print("\n Update Product Quantity ")
    
    if not inventory:
        print("Inventory is empty! Please add products first.")
        return
    
    product_name = input("Enter product name to update: ").strip().title()
    
    # Check if product exists
    if product_name not in inventory:
        print(f"Sorry: '{product_name}' not found in inventory!")
        return
    
    try:
        new_quantity = int(input(f"Enter new quantity for '{product_name}': "))
        if new_quantity < 0:
            print("Error: Quantity cannot be negative!")
            return
        
        old_quantity = inventory[product_name]        #Updating from old quantity to new quantity
        inventory[product_name] = new_quantity
        print(f" '{product_name}' quantity updated from {old_quantity} to {new_quantity}")
    except ValueError:
        print("Error: Please enter a valid number for quantity!")
```

```python
update_quantity()
```

```python
#Search for a product by name

def search_product():
    print("\n Search Product ")
    
    if not inventory:
        print("Inventory is empty!")
        return
    
    product_name = input("Enter product name to search: ").strip().title()
    
    # Check if product exists
    if product_name in inventory:
        print(f" Found: '{product_name}' - Quantity: {inventory[product_name]}")
    else:
        print(f" '{product_name}' not found in inventory!")
        
        # Optional: Show similar products
        similar_products = [p for p in inventory.keys() if product_name.lower() in p.lower()]
        if similar_products:
            print("\n Did you mean one of these?")
            for product in similar_products[:5]:  # Show up to 5 similar products
                print(f"  - {product}")
```

```python
#To remove a product from the inventory

def remove_product():
    print("\n Remove Product ")
    
    if not inventory:
        print("Inventory is empty!")
        return
    
    product_name = input("Enter product name to remove: ").strip().title()
    
    # Check if product exists
    if product_name in inventory:
        confirm = input(f"Are you sure you want to remove '{product_name}'? (yes/no): ").lower()
        if confirm == 'yes' or confirm == 'y':
            removed_quantity = inventory.pop(product_name)
            print(f" Alright!'{product_name}' with quantity {removed_quantity} has been removed!")
        else:
            print("Removal cancelled.")
    else:
        print(f"Error: '{product_name}' not found in inventory!")
```

##### Low Stock Report
We assume that when any of the stock gets to 5, our stock is low. So we display all products with quantity less than 5

```python
#Low stock report

def low_stock_report():
    print("\n Low Stock Report (Quantity < 5)")
    
    low_stock_items = {product: qty for product, qty in inventory.items() if qty < 5}
    
    if not low_stock_items:
        print(" All products have sufficient stock (5 or more units)!")
        return
    
    print("\n LOW STOCK! Please Restock!")
    print("{:<30} {:>10}".format("Product Name", "Quantity"))
    print("-" * 42)
    
    # Sort by quantity (lowest first)
    for product, quantity in sorted(low_stock_items.items(), key=lambda x: x[1]):
        print("{:<30} {:>10}".format(product, quantity))
    print("-" * 42)
    print(f"Total low stock items: {len(low_stock_items)}")
    
    # Additional alert for critical stock (quantity = 0)
    critical_items = [p for p, q in low_stock_items.items() if q == 0]
    if critical_items:
        print("\n CRITICAL: Items with NO Available Stocks!!:")
        for item in critical_items:
            print(f"  - {item}")
```

```python
#Ti display the full inventory analytics

def inventory_statistics():
    """Display inventory statistics"""
    print("\n Inventory Statistics ")
    
    if not inventory:
        print("No data available. Inventory is empty!")
        return
    
    # Total number of products
    total_products = len(inventory)
    
    # Total stock available
    total_stock = sum(inventory.values())
    
    # Product with highest quantity
    max_product = max(inventory.items(), key=lambda x: x[1])
    
    # Product with lowest quantity
    min_product = min(inventory.items(), key=lambda x: x[1])
    
    # Calculate average stock
    average_stock = total_stock / total_products if total_products > 0 else 0
    
    # Display statistics
    print("\n STATISTICS SUMMARY")
    print("="*40)
    print(f" Total number of products: {total_products}")
    print(f" Total stock available: {total_stock} units")
    print(f" Average stock per product: {average_stock:.1f} units")
    print(f" Product with highest quantity: '{max_product[0]}' ({max_product[1]} units)")
    print(f" Product with lowest quantity: '{min_product[0]}' ({min_product[1]} units)")
    print("="*40)
    
    # Additional insight: Show distribution
    print("\n Stock Distribution:")
    print(f"  • Products with 0 stock: {len([q for q in inventory.values() if q == 0])}")
    print(f"  • Products with 1-4 stock: {len([q for q in inventory.values() if 1 <= q < 5])}")
    print(f"  • Products with 5-19 stock: {len([q for q in inventory.values() if 5 <= q <= 19])}")
    print(f"  • Products with 20+ stock: {len([q for q in inventory.values() if q >= 20])}")
```
